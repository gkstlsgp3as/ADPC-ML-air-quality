```python
def csv_from_excel(path,sheet_l,file):
    for i,sheet in enumerate(sheet_l): 
        if file != 0:
            sh=wb_data.sheet_by_name(sheet)
            csv_file=open(path+file[4:7]+'_'+str(len(sheet_l))+'-'+str(i+1)+'.csv','w',newline='',encoding='utf-8')
            
            wr=csv.writer(csv_file,quoting=csv.QUOTE_ALL)
            for i,rownum in enumerate(range(sh.nrows)):
                if i < 2: 
                    wr.writerow(sh.row_values(rownum))
                if type(sh.row_values(rownum)[0]) != type('string'):
                    wr.writerow(sh.row_values(rownum))
        else: 
            sh=wb.sheet_by_name(sheet)
            csv_file=open(path+sheet[0:3]+'_19-20'+'.csv','w',newline='',encoding='utf-8')
            
            wr=csv.writer(csv_file,quoting=csv.QUOTE_ALL)
            for i,rownum in enumerate(range(sh.nrows)):
                if i == 0: 
                    wr.writerow(sh.row_values(rownum))
                if type(sh.row_values(rownum)[0]) != type('string'):
                    wr.writerow(sh.row_values(rownum))
        
        csv_file.close()
```

##### processing data into a dataframe with designed structure

```python
path='./ML/Hourly_Data_csv/'
path2='./ML/Hourly_Data_csv_19-20/'
file_list=os.listdir(path)
file_list2=os.listdir(path2)
print(file_list[:10])
print(file_list2[:10])
```

```python
for i, file in enumerate(file_list):
    with open(path+file) as infile:
        reader = csv.reader(infile)
        headers = next(reader)
        
    #if column is empty, ignore it to prevent 'unnamed' columns
    header_indices = [i for i, item in enumerate(headers) if item] 
    #print(i, headers)
    if i != 0: 
        df_samp=pd.read_csv(path+file, usecols=header_indices)
         
        df_samp['stationid']=str(file[0:3])
        
        for col in df_samp.columns:
            if col in [' Rel hum', ' Rel hum.1', 'RH']:
                if df_samp[col][0].startswith('at 2 m'):
                    df_samp=df_samp.rename(columns={col:' Rel hum'})
                elif df_samp[col][0].startswith('at 10 m'):
                    df_samp=df_samp.rename(columns={col:' Rel hum_10m'})
                elif df_samp[col][0].startswith('at 30 m'):
                    df_samp=df_samp.rename(columns={col:' Rel hum_30m'})
                elif df_samp[col][0].startswith('at 3 m'):
                    df_samp=df_samp.rename(columns={col:' Rel hum_3m'})
            elif col in ['PM10', 'PM10(µg./m3)']: 
                df_samp=df_samp.rename(columns={col:' PM10'})
            elif col == 'PM2.5': 
                df_samp=df_samp.rename(columns={col:' PM2.5'})
            elif col == 'NRAD': 
                df_samp=df_samp.rename(columns={col:' Net rad'})
            elif col in ['Rain', 'RAIN']: 
                df_samp=df_samp.rename(columns={col:' Rain'})
            elif col == 'BP': 
                df_samp=df_samp.rename(columns={col:' Pressure'})
            elif col in ['Glob rad', 'SRAD']: 
                df_samp=df_samp.rename(columns={col:' Glob rad'})
            elif col == 'YY/MM/DD.1': 
                df_samp=df_samp.rename(columns={col:'YY/MM/DD'})
            elif col in ['WS', ' WS ']: 
                df_samp=df_samp.rename(columns={col:' Wind speed'})
            elif col in [' Temp',' TEMP.1', 'TEMP', 'TMP', 'Temp', ' Temp.1']: 
                if df_samp[col][0].startswith('at 2 m'):
                    df_samp=df_samp.rename(columns={col:' Temp'})
                elif df_samp[col][0].startswith('at 10 m'):
                    df_samp=df_samp.rename(columns={col:' Temp_10m'})
                elif df_samp[col][0].startswith('at 30 m'):
                    df_samp=df_samp.rename(columns={col:' Temp_30m'})
                elif df_samp[col][0].startswith('at 3 m'):
                    df_samp=df_samp.rename(columns={col:' Temp_3m'})
            elif col == 'WD': 
                df_samp=df_samp.rename(columns={col:' Wind dir'})
        pcd_df=pd.concat([pcd_df,df_samp[1:]],ignore_index=True)
        
    else: 
        pcd_df = pd.read_csv(path+file, usecols=header_indices)
        pcd_df['stationid']=str(file[0:3])
        
        for col in pcd_df.columns:
            if col in [' Rel hum', ' Rel hum.1', 'RH']: 
                if pcd_df[col][0].startswith('at 2 m'):
                    pcd_df=pcd_df.rename(columns={col:' Rel hum'})
                elif pcd_df[col][0].startswith('at 10 m'):
                    pcd_df=pcd_df.rename(columns={col:' Rel hum_10m'})
                elif pcd_df[col][0].startswith('at 30 m'):
                    pcd_df=pcd_df.rename(columns={col:' Rel hum_30m'})
                elif pcd_df[col][0].startswith('at 3 m'):
                    pcd_df=pcd_df.rename(columns={col:' Rel hum_3m'})
            elif col in ['PM10', 'PM10(µg./m3)']:  
                pcd_df=pcd_df.rename(columns={col:' PM10'})
            elif col == 'PM2.5': 
                pcd_df=pcd_df.rename(columns={col:' PM2.5'})
            elif col == 'NRAD': 
                pcd_df=pcd_df.rename(columns={col:' Net rad'})
            elif col in ['Rain', 'RAIN']: 
                pcd_df=pcd_df.rename(columns={col:' Rain'})
            elif col == 'BP': 
                pcd_df=pcd_df.rename(columns={col:' Pressure'})
            elif col in ['Glob rad', 'SRAD']: 
                pcd_df=pcd_df.rename(columns={col:' Glob rad'})
            elif col == 'YY/MM/DD.1': 
                pcd_df=pcd_df.rename(columns={col:'YY/MM/DD'})
            elif col in ['WS', ' WS ']:  
                pcd_df=pcd_df.rename(columns={col:' Wind speed'})
            elif col in [' Temp',' TEMP.1', 'TEMP', 'TMP', 'Temp', ' Temp.1']: 
                if pcd_df[col][0].startswith('at 2 m'):
                    pcd_df=pcd_df.rename(columns={col:' Temp'})
                elif pcd_df[col][0].startswith('at 10 m'):
                    pcd_df=pcd_df.rename(columns={col:' Temp_10m'})
                elif pcd_df[col][0].startswith('at 30 m'):
                    pcd_df=pcd_df.rename(columns={col:' Temp_30m'})
                elif pcd_df[col][0].startswith('at 3 m'):
                    pcd_df=pcd_df.rename(columns={col:' Temp_3m'})    
            elif col == 'WD': 
                pcd_df=pcd_df.rename(columns={col:' Wind dir'})
        pcd_df=pcd_df.drop(index=0)
        
pcd_df=pcd_df.rename(columns={"YY/MM/DD":'date'})
pcd_df=pcd_df.rename(columns={"Hour":'time'})
```

```python
#format the 'date' column as YYYY-MM-DD
pcd_df['date']='20'+pcd_df['date'].astype('int').astype('str')
#format the 'time' column as HH:MM:SS
pcd_df['time']=((pcd_df['time'].astype('int')//100)-1).astype(str)
pcd_df['time']=pcd_df['time'].apply(lambda _: _.zfill(2))+':00:00'
#concat two strings from each column
pcd_df.insert(0,'datetime',pcd_df['date']+" "+pcd_df['time'])
```

```python
#format to the datetime type
pcd_df['datetime'] = pcd_df['datetime'].apply(lambda _: dt.strptime(_,"%Y%m%d %H:%M:%S"))
```
